```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
# Assuming your file is named con.out and is in the same directory as your Python script
file_path = 'ce_cross11.out'

# Read data from the file
with open(file_path, 'r') as file:
    lines = file.readlines()

# Extract data from the file (taking only the first two columns)
x_values = []
y_values = []

for line in lines:
    # Assuming columns are space-separated
    columns = line.split()
    if len(columns) >= 2:
        x_values.append(float(columns[0]))
        y_values.append(float(columns[2]))

# Plot the data
y_values=np.array(y_values)*10**(-24)
x_values=x_values[200:700]
y_values=y_values[200:700]

plt.plot(x_values, y_values, label='Data from con.out')
plt.plot(x_values, y_values, '-')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Your Plot Title')


# Read the Excel file
df = pd.read_excel('3s_3p.xlsx')

# Assuming column names are 'Column1' and 'Column2', replace them with your actual column names
column1 = df['C1']
column2 = df['C2']
# Set y-axis in scientific notation with a fixed power of 10


# Plot the graph
plt.plot(column1, column2,'.')
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.title('Your Title')
plt.grid(True)
plt.legend()
plt.show()
```
